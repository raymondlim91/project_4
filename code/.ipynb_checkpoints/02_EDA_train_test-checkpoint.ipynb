```python
# the BBox for the map provided in the asset folder is (-88, -87.5, 41.6, 42.1)

# check if the range of coordinates are within the map provided 
BBox_spray = ((df_spray.longitude.min(), df_spray.longitude.max(),
        df_spray.latitude.min(), df_spray.latitude.max()))

BBox_spray
```

```python
# drop sprays that took place at coordinates outside the map provided
df_spray = df_spray.loc[df_spray["latitude"] < 42.1]
df_spray.info()
```

```python
virus_present = df_train[(df_train["year"]==2011) | (df_train["year"]==2013) ]
virus_present = virus_present[virus_present["WnvPresent"]==1]
```

```python
mapdata = np.loadtxt("../assets/mapdata_copyright_openstreetmap_contributors.txt")

aspect = mapdata.shape[0] * 1.0 / mapdata.shape[1]
lon_lat_box = (-88, -87.5, 41.6, 42.10)

plt.figure(figsize=(10,14))
plt.imshow(mapdata, 
           cmap=plt.get_cmap('gray'), 
           extent=lon_lat_box, 
           aspect=aspect)


spray_2011 = df_spray.loc[df_spray["year"]==2011]
spray_2013 = df_spray.loc[df_spray["year"]==2013]


sp_2011 = plt.scatter(spray_2011['longitude'], spray_2011['latitude'], c="orange", marker='o', s=1, alpha=1)
sp_2013 = plt.scatter(spray_2013['longitude'], spray_2013['latitude'], c="brown", marker='o', s=1, alpha=1)
virus_present = plt.scatter(virus_present['Longitude'], virus_present["Latitude"], c = "blue", marker = 'o', s=1, alpha = 1)

plt.title("Location of Sprays in 2011 (yellow) and 2013 (red)", fontsize=15)
plt.legend((sp_2011, sp_2013 ), ("yellow: 2011", "red: 2013"))
plt.show();
```

### Count of mosquitos and WnvPresent by Species

```python
## To datetime
df_train['Date']=pd.to_datetime(df_train['Date'])
df_weather['Date']=pd.to_datetime(df_weather['Date'])
```

```python
df_tr_species =  df_train.groupby('Species')['NumMosquitos', 'WnvPresent'].sum()
```

```python
df_tr_species.sort_values('NumMosquitos', inplace=True)
```

```python
plt.barh(df_tr_species.index, df_tr_species['NumMosquitos'], label='NumMosquitos')
plt.title("Mosquito count by Species", fontdict={'size': 15});
```

```python
plt.barh(df_tr_species.index, df_tr_species['WnvPresent'], label='WnvPresent');
plt.title("West Nile Virus by Species", fontdict={'size': 15});
```

### Count of mosquitos and WnvPresent by Trap

```python
df_tr_trap  = df_train.groupby('Trap')['NumMosquitos', 'WnvPresent'].sum()
```

```python
df_tr_trap_top10 = df_tr_trap.sort_values('NumMosquitos', ascending=True).tail(10)
```

```python
df_tr_trap_top10
```

```python
plt.barh(df_tr_trap_top10.index, df_tr_trap_top10['NumMosquitos'], label='NumMosquitos')
plt.title("Mosquito count by Trap", fontdict={'size': 15});
```

```python
plt.barh(df_tr_trap_top10.index, df_tr_trap_top10.sort_values('WnvPresent')['WnvPresent'], label='WnvPresent')
plt.title("West Nile Virus by Trap", fontdict={'size': 15});
```
