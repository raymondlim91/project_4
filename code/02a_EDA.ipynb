### Count of mosquitos and WnvPresent by Species

```python
## To datetime
df_train['Date']=pd.to_datetime(df_train['Date'])
df_weather['Date']=pd.to_datetime(df_weather['Date'])
```

```python
## Filter for only WNV positive
df_train_wnvpresent = df_train.loc[df_train["WnvPresent"]==1]
```

```python
## Show number of Mosquitos by species
a = df_train_wnvpresent.groupby("Species")["NumMosquitos"].sum()/df_train_wnvpresent["NumMosquitos"].sum()
```

```python
a = pd.DataFrame(a)
```

```python
## Show number of Mosquitos by species with WNV positive
b = df_train.groupby("Species")["NumMosquitos"].sum()/df_train["NumMosquitos"].sum()
```

```python
b = pd.DataFrame(b)
```

```python
ax = sns.barplot(data = a, x=a.index, y="NumMosquitos")
plt.xticks(rotation = 80);
```

**Interpretation** - Most of the species are Culex Pipiens and Culex Restuans. 

```python
ax1 = sns.barplot(data = b, x = b.index, y = "NumMosquitos")
plt.xticks(rotation = 80);
```

**Interpretation** - For WNV present, most of the species are also Culex Pipiens and Culex Restuans. 

```python
df_tr_species =  df_train.groupby('Species')['NumMosquitos', 'WnvPresent'].sum()
```

```python
df_tr_species
```

```python
df_tr_species.sort_values('NumMosquitos', inplace=True)
```

```python
plt.barh(df_tr_species.index, df_tr_species['NumMosquitos'], label='NumMosquitos')
plt.title("Mosquito count by Species", fontdict={'size': 15});
```

```python
plt.barh(df_tr_species.index, df_tr_species['WnvPresent'], label='WnvPresent');
plt.title("West Nile Virus by Species", fontdict={'size': 15});
```

### Count of mosquitos and WnvPresent by Trap

```python
df_tr_trap  = df_train.groupby('Trap')['NumMosquitos', 'WnvPresent'].sum()
```

```python
df_tr_trap_top10 = df_tr_trap.sort_values('NumMosquitos', ascending=True).tail(10)
```

```python
df_tr_trap_top10
```

```python
plt.barh(df_tr_trap_top10.index, df_tr_trap_top10['NumMosquitos'], label='NumMosquitos')
plt.title("Mosquito count by Trap", fontdict={'size': 15});
```

```python
plt.barh(df_tr_trap_top10.index, df_tr_trap_top10.sort_values('WnvPresent')['WnvPresent'], label='WnvPresent')
plt.title("West Nile Virus by Trap", fontdict={'size': 15});
```
